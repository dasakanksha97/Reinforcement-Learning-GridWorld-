```python
import numpy as np
import matplotlib.pyplot as plt
import gym
from gym import spaces
```

```python
class GridEnvironment(gym.Env):
    metadata = {'render.modes' : []} 
    
    def __init__(self):
        self.observation_space = spaces.Discrete(25)
        self.action_space = spaces.Discrete(5)
        self.max_timesteps = 10 
    
    def reset(self):
        self.timestep = 0
        self.agent_pos = [4,0]
        self.goal_pos = [0,4]
        self.state = np.zeros((5,5))
        self.state[tuple(self.agent_pos)] = 1
        self.state[tuple(self.goal_pos)] = 2
        observation = self.state.flatten()
        return observation
    
    def step(self, action):
        self.state = np.random.choice(self.observation_space.n)
        reward = 0
        if action == 0: #Moves down by a step
            self.agent_pos[0] += 1
        if action == 1:  #Moves up by a step
            self.agent_pos[0] -= 1 
        if action == 2:  #Moves right by a step
            self.agent_pos[1] += 1      
        if action == 3:  #Moves left by a step
            self.agent_pos[1] -=1 

        if (self.agent_pos[0] == 0 and self.agent_pos[1]==0):
              reward =5
              print(f"positive battery {reward}")
        elif(self.agent_pos[0]==0 and self.agent_pos[1]==3):
              reward =6
              print(f"positive battery {reward}")
        elif(self.agent_pos[0]==4 and self.agent_pos[1]==3):
              reward =-5
              print(f'negative rock {reward}')
        elif(self.agent_pos[0]==3 and self.agent_pos[1]==4):
              reward =-6
              print(f"negative rock {reward}")
        self.agent_pos = np.clip(self.agent_pos, 0, 4) #Clips the values
        self.state = np.zeros((5,5))
        self.state[tuple(self.agent_pos)] = 1
        self.state[tuple(self.goal_pos)] = 2
        observation = self.state.flatten()
        
        
        if(self.agent_pos == self.goal_pos).all(): 
            reward= 1
        
        self.timestep +=1
        done = True if self.timestep >= self.max_timesteps else False
        #done = True if reward == 1 else False
        info = {}
        
        return observation, reward, done, info
    
    def render(self):
        plt.imshow(self.state)
            
            
```

```python
env = GridEnvironment()
```

```python
#Tuning parameters definition
Q_table= np.zeros((env.observation_space.n, env.action_space.n))

gamma = 0.8
epsilon = 1.5
alpha_value = 0.7
number_total_episodes = 5000
number_of_iterations = 50
reward=0
decay_rate = 0.001
min_epsilon = 0.1

class RandomAgent:
    def __init__(self,env):
        self.observation_space = env.observation_space
        self.action_space = env.action_space
        
    def step(self,state, epsilon):
        state-=1
        if np.random.uniform(0,1)<=epsilon:
           return np.random.choice(self.action_space.n)
        else:
           return np.argmax(Q_table[state, :])

#Updation formula for Q table
def SARSA(s_t, s_tp1, a_t, a_tp1, reward):
      s_t=s_t-1
      s_tp1-=1
      
      Q_table[s_t,a_t] = Q_table[s_t, a_t]+ alpha_value * ((reward + gamma * Q_table[s_tp1, a_tp1]) - (Q_table[s_t, a_t]))

obj = RandomAgent(env)
episodes=[]
epsilons = []
r=[]
for i in range(number_total_episodes):    
    s_t = env.reset()
    count = 0
    total_reward=0
    epsilon = epsilon- (epsilon * decay_rate)  #, min_epsilon)
    if epsilon<min_epsilon:
      epsilon=min_epsilon
    a_t = obj.step(s_t, epsilon)
    
    while count < number_of_iterations:
        env.render()
        s_tp1, reward, done, info = env.step(a_t) 
        a_tp1 = obj.step(s_tp1,epsilon)
        SARSA(s_t, s_tp1,a_t, a_tp1,reward) 
        s_t = s_tp1
        a_t = a_tp1
        count += 1
        total_reward += reward
        
        if done:
           
            episodes.append(i+1)
            epsilons.append(epsilon)
            
            r.append(reward)
            break






```
